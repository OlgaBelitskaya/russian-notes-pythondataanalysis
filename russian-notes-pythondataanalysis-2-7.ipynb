# 2.7 Преобразование типов данных в Python
## Базовые объектные типы

```python
# отсутствие значений
print(type(None))
# логические
print(type(True), type(False), type(False and 0))
# целочисленные
print(type(0), type(0 & 1), type(False and 1))
# вещественные и комплексные
print(type(5.4), type(5 + 4j), type(complex(5)))
# списки и кортежи
print(type([1, 5.3, False, 4]), type((1, True, 3, 5+4j)))
# интервалы, списки, кортежи
print(type(range(5)), type(list(range(5))), type(tuple(range(5))))
# строковые
print(type('Hello'), type(chr(155)))
print(type(bin(12)), type(hex(12)))
# байтовые
print(type(b'a'), type(bytearray([1,2,3])))
print(type(memoryview(bytearray('XYZ', 'utf-8'))))
# изменямые и неизменяемые множества
print(type({'a', 3, True}), type(frozenset({False, 2, 'r'})))
# словари
print(type({'a': 1000}), type({}))
```

## Простые встроенные функции

[Простые встроенные функции - документация Python](https://docs.python.org/3/library/functions.html)


Приведите примеры использования встроенных функции 

для конвертации одного типа данных `Python` в другой.

```python
# abs() не меняет тип
a = 0b101
print(a)
print(type(a))
print(abs(a))
print(type(abs(a)))
# abs() меняет тип
a = 5 + 2j
print(a)
print(type(a))
print(abs(a))
print(type(abs(a)))
```

```python
# int() меняет тип
x = 2e-3
print(x)
print(x+.9,int(x+.9))
```

```python
# hex() меняет тип на строковый (!!!)
x = 2000
print(x)
print(hex(x))
print(type(hex(x)))
```

```python
# tuple() меняет тип
lst = [1, 2, 3, 4]
print(type(lst))
print(type(tuple(lst)))
```

```python
# применение к строковым литералам
bool('hello'), list('hello'), tuple('hello'), set('hello')
```

```python
# map() не меняет тип (остается список)), 
# но меняет тип элементов списка
list(map(int,['1','2','3']))
```

```python
# map() меняет тип одного значение
for el in map(int,'1'):
    print(el)
```

```python
# sorted() меняет тип на список
s1={1, 4, 3, 6, 2, 8, 11, 32}
print(sorted(s1))
```

```python
# значения одинаковые, типы разные
x = 1.0; y = 1
x == y, hash(x) == hash(y), type(x) == type(y)
```
