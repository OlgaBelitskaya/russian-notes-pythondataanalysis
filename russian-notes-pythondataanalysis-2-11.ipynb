# 2.11 Управляющяя конструкция исключения
Основные управляющие конструкции языка:
- условия `if ... elif ... else ...`
- циклы `for ... in ...`, `while ...`
- исключения `try ... except ...`

Ключевые слова
- `assert`
- `raise`
- `try`
- `except`
- `else`
- `finally`

```python
try:
    x,y=1,0
    assert y!=0,"Деление на ноль невозможно!"
    print(x / y)
except AssertionError as msg:
        print(msg) 
print("Дальше программа все равно работает!")
```

```python
# является ли число целым и нечетным
odd_lambda = lambda x: int(x) if (x%2!=0) and (int(x)==x) else None
# ошибка, если другого вида 
def odd_assert(x):
    assert((x%2!=0) and (int(x)==x))
    return x
odd_lambda(3.0),odd_lambda(3.1),odd_assert(3)
```

```python
# можно ли совместить lambda и assert?
def assert1(cond): assert cond
def assert2(cond): 
    try:
        assert cond
    except AssertionError:
        print("the assertion error!")
# ошибка, если значения не совпадают
eq1 = lambda x,y: assert1(x == y)
eq2 = lambda x,y: assert2(x == y)
print(20*'=')
eq1([1,0],[1.,0.])
print(20*'=')
eq2([1,0],[2.,0.])
```

```python
# конструкция с else
x=y=10
try:
    assert(x==y)
except AssertionError as error:
    print(error)
else:
    print('запуск команд после else')
```

```python
# конструкция с else, finally
x,y=10,10.5
try:
    assert(x==y)
except AssertionError as error1:
    print('числа не равны')
else:
# не проверяется в случае неравенства
    try:
        assert(x==int(x))
        assert(y==int(y))
    except AssertionError as error2:
        print('числа не являются целыми')
finally:
    print('обнаруженные исключения распечатаны')
```

```python
def try_test(id):
    try:
        print('запуск try блока =>')
        lst=[1,2,3]
        lst[id]
    except Exception as e:
        print('запуск exception блока => ',e)
    else:
        print('запуск else блока => ')
    finally:
        print('запуск finally блока => ')
try_test(1); print(50*'='); try_test(5)
```

```python
# вместе с условием if
x = 10
exc_str = f'x должно быть <= 5, значение x было равно {x}'
if x > 5:
    try:  
        raise Exception
    except Exception as error:
        print(exc_str)
else:
    print(x)
```

```python
# построение пользовательских исключений
class UserError1(Exception):
    def __init__(self, value):
        self.value = value
    def __str__(self):
        return(repr(self.value))
# любое численное значение 
try:
    raise UserError1(10e12) 
# значение установленного исключения - в ошибках
except UserError1 as error:
    print(f'пользовательское исключение 1: {error.value}')
# строковые значения
class UserError2(RuntimeError):
    def __init__(self, arg):
        self.args = arg
try:
    raise UserError2("error")
except UserError2 as error:
    print(f'пользовательское исключение 2: {error.args}')
```

```python
class Error(Exception):
    """базовый пользовательский класс"""
    pass
class ZeroValue(Error):
    """проверка на равенство нулю"""
    pass
def check_number(n):
    try:
        if n != 0: return n
        else: raise ZeroValue
    except ZeroValue:
         print("Значение аргумента равно 0, выберите другое!")
check_number(0)
check_number(5)
```

```python
class Error(Exception):
    # базовый класс
    pass
class UserError3(Error):
    # недопустимые значения
    def __init__(self, msg, prev, next):
        self.prev = prev
        self.next = next
        # сообщение об ошибке
        self.msg = msg
try:
    raise UserError3('недопустимые значения',-10**21,10e20) 
except UserError3 as error:
    print(f'{error.msg}: {error.prev,error.next}')
```
