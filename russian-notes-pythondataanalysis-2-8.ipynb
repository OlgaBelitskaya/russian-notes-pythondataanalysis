```python
help('FORMATTING')
```

```python
# —Å—Ç—Ä–æ–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–æ –ø–æ —Ü–µ–Ω—Ç—Ä—É
# —Å–∏–º–≤–æ–ª –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è ‚ôô
print('{:‚ôô^35}'.format('–ø–æ —Ü–µ–Ω—Ç—Ä—É'))
print('{:35}'.format('–ø–æ –ª–µ–≤–æ–º—É –∫—Ä–∞—é –±–µ–∑ –∑–∞–ø–æ–ª–Ω–∏—Ç–µ–ª—è'))
print('{:‚ôô<35}'.format('–ø–æ –ª–µ–≤–æ–º—É –∫—Ä–∞—é'))
print('{:‚ôô>35}'.format('–ø–æ –ø—Ä–∞–≤–æ–º—É –∫—Ä–∞—é'))
```

```python
# —Å—Ä–µ–∑ –Ω–∞—á–∞–ª–∞ —Å—Ç—Ä–æ–∫–∏ –∏–∑ 5 –±—É–∫–≤,
# –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ | –¥–æ 20 —Å–∏–º–≤–æ–ª–æ–≤ –∏ –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏–µ –ø–æ —Ü–µ–Ω—Ç—Ä—É
print("{:|^20.5}.".format("‚ôî‚ôï‚ôñ‚ôó‚ôò‚ôô"))
```

```python
# —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã
for x in range(1, 11):
    print(f'{x:02} {x*x:3} {x*x*x:4}')
```

```python
x = 12.333
# –ø–æ–∫–∞–∑–∞—Ç—å –∑–Ω–∞–∫–∏ +/-
print("{:+f} {:+f}".format(x, -x))
# –ø–æ–∫–∞–∑–∞—Ç—å —Ç–æ–ª—å–∫–æ –∑–Ω–∞–∫–∏ -
print("{:-f} {:-f}".format(x, -x))
# –ø–æ–∫–∞–∑–∞—Ç—å –ø—Ä–æ–±–µ–ª –≤–º–µ—Å—Ç–æ –∑–Ω–∞–∫–∞ +
print("{: f} {: f}".format(x, -x))
```

```python
# –∏–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã format()
print("X = {}. Y = {b}. Z = {a}.".format(1, a=3, b=2))
```

```python
# —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É, 
# –∏—Å–ø–æ–ª—å–∑—É—è join(), format()
fruits=['üçä','üçã','üçé','üçê']
print(' | {} | '.join(fruits).format(1,2,3))
```

```python
# –¥–≤–æ–∏—á–Ω—ã–π, –≤–æ—Å—å–º–µ—Ä–∏—á–Ω—ã–π –∏ —à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç —á–∏—Å–ª–∞
print("bin: {0:b}, oct: {0:o}, hex: {0:x}".format(128))
```

```python
# –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∏–π —à–∞–±–ª–æ–Ω —Ñ–æ—Ä–º–∞—Ç–∞ str
string = "{:{fill}{align}{width}.{cut}}"
print(string.format(
    ' üçä !!!', fill='üçã', align='^', width=15, cut=3))
# –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∏–π —à–∞–±–ª–æ–Ω —Ñ–æ—Ä–º–∞—Ç–∞ float
number = "{:{fill}{align}{width}.{precision}f}"
print(number.format(
    1.23456789, fill='|', align='>', width=31, precision=6))
```

- `ANSI Escape Sequences / Codes`
   - `\033[` = escape code (–¥—Ä—É–≥–∏–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã: `\x1b[`,`\u001b[`)
   - `style`+ `;` + `color` + `;` + `background color` +`m`

```python
esc = "\u001b["
style_and_color_and_back = "3;35;43m "
string = "—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞"
print(esc+style_and_color_and_back+string)
```

- –∏–ª–∏
    - `\033[48;5;` + background color +`m` + ` `+ `\033[38;5;` + color +`m`

```python
def c256(st,c,bc,n):
    back = f"\033[48;5;{str(bc)}m"
    color = f"\033[38;5;{str(c)}m"
    string = f"{str(st):{int(n)}s}"
    return back + " " + color + " " + string + " \033[0;0m"
print(' '+' '.join(
    [c256(x,x,253,3) + '\t' + (x % 8 == 7) * '\n' 
     for x in range(256)]))
```

```python
for el in ['a', 'abcd']:
    print(c256(el,78,191,len(el)+1), end=" ")
```
