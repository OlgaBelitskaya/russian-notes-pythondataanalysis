### Получение данных о точном времени с помощью модулей Python

````python
from time import gmtime, strftime
print('Точное время')
gmt_now = strftime('%A %Y-%m-%d %H:%M:%S %Z', gmtime())
gmt_now
````

````python
import datetime, pytz
now = datetime.datetime.now(pytz.timezone('Europe/Moscow'))
print('Точное время с учетом гепозиции')
now = now.strftime('%A %Y/%m/%d %H:%M:%S %Z')
now
````

````python
#описание текущей даты
import datetime
today = datetime.date.today()
print ('сегодня:\n', today)
print ('с указанием дня недели:\n', today.ctime())
print ('кортеж:\n', today.timetuple())
print ('номер дня по Григорианскому календарю:\n', today.toordinal())
print ('год:\n', today.year)
print ('месяц:\n', today.month)
print ('день :\n', today.day)
````

````python
import numpy as np
print('Точное время NumPy')
str(np.datetime64('now')).replace('T',' ')
````

````python
import pandas as pd
print('Точное время Pandas')
print(str(pd.to_datetime('now')))
````

### Отображение промежутка времени

````python
import ipywidgets as iw, pandas as pd, datetime as dt
start_date = dt.datetime.today()
end_date = dt.datetime(2022, 11, 18)
dates = pd.date_range(start_date, end_date, freq='D')
options = [(date.strftime(' %d %b %Y '), date) for date in dates]
index = (0, len(options)-1)
selection_range_slider = iw.SelectionRangeSlider(
    options=options, index=index, 
    description='онлайн-курс', orientation='horizontal',
    layout={'width': '700px'})
selection_range_slider
````

### Интерактивные инструменты

````python
#@title Ввод текста {run:"auto"}
# Использована опция ячейки
# "Add a form / Добавить форму"
# строковые переменные
text = 'The quick brown fox jumps over the lazy dog.' #@param {type:"string"}
word = 'dog' #@param {type:"string"}
````

````python
# интерактивные данные
import re
match=re.search(word, text)
s=match.start(); e=match.end()
print(word + '\n' + text)
print('Found "%s" in "%s" from %d to %d '%\
    (match.re.pattern, match.string, s, e))
````

````python
# built-in функция input()
import re
pattern=input()
text=input()
match=re.search(pattern, text)
s=match.start()
e=match.end()
print('Found "%s" in "%s" from %d to %d '%\
    (match.re.pattern,match.string,s,e))
````
