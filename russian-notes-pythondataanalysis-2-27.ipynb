### –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–µ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã

````python
#@title –í–≤–æ–¥ —Ç–µ–∫—Å—Ç–∞ {run:"auto"}
# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∞ –æ–ø—Ü–∏—è —è—á–µ–π–∫–∏
# "Add a form / –î–æ–±–∞–≤–∏—Ç—å —Ñ–æ—Ä–º—É"
# —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
text = 'The quick brown fox jumps over the lazy dog.' #@param {type:"string"}
word = 'dog' #@param {type:"string"}
````

````python
# –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
import re
match=re.search(word, text)
s=match.start(); e=match.end()
print(word + '\n' + text)
print('Found "%s" in "%s" from %d to %d '%\
    (match.re.pattern, match.string, s, e))
````

````python
# built-in —Ñ—É–Ω–∫—Ü–∏—è input()
import re
pattern=input()
text=input()
match=re.search(pattern, text)
s=match.start()
e=match.end()
print('Found "%s" in "%s" from %d to %d '%\
    (match.re.pattern,match.string,s,e))
````
